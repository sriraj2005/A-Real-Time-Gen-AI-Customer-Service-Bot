## Sentiment Analysis for Chatbot

First, we need to install the `transformers` library, which provides access to pre-trained models for various NLP tasks, including sentiment analysis.

```python
pip install transformers
```

Now, let's load a pre-trained sentiment analysis model from Hugging Face. We'll use a pipeline for easy inference.

```python
from transformers import pipeline

# Load a pre-trained sentiment analysis model
sentiment_analyzer = pipeline('sentiment-analysis')

print("Sentiment analysis model loaded successfully!")
```

Here's a helper function to analyze the sentiment of a given text and extract the sentiment label and score.

```python
def get_sentiment(text):
  """Analyzes the sentiment of a given text using the pre-trained model."""
  result = sentiment_analyzer(text)[0]
  return result['label'], result['score']

# Example usage:
text1 = "I love this new feature! It's amazing."
label1, score1 = get_sentiment(text1)
print(f"Text: '{text1}'\nSentiment: {label1} (Score: {score1:.2f})\n")

text2 = "This service is terrible and I'm very disappointed."
label2, score2 = get_sentiment(text2)
print(f"Text: '{text2}'\nSentiment: {label2} (Score: {score2:.2f})\n")

text3 = "The weather is neither good nor bad today."
label3, score3 = get_sentiment(text3)
print(f"Text: '{text3}'\nSentiment: {label3} (Score: {score3:.2f})\n")
```
