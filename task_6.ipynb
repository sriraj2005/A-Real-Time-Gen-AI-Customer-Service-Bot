```python
from langdetect import detect, DetectorFactory

# Ensure consistent results across runs
DetectorFactory.seed = 0

def get_chatbot_response(user_input):
    """
    Detects the language of the user input and provides a basic response.
    """
    try:
        detected_lang = detect(user_input)
    except Exception:
        # Fallback if language detection fails (e.g., very short input)
        detected_lang = 'en'

    # Define simple responses for demonstration
    responses = {
        'en': "Hello! How can I help you today?",
        'es': "ВЎHola! ВїCГіmo puedo ayudarte hoy?",
        'fr': "Bonjour! Comment puis-je vous aider aujourd'",
        'de': "Hallo! Wie kann ich Ihnen heute helfen?",
        'it': "Ciao! Come posso aiutarti oggi?"
    }

    # Provide a generic response if the language is not explicitly supported
    response = responses.get(detected_lang, f"I detect you're speaking {detected_lang}. Unfortunately, I only have pre-defined responses in English, Spanish, French, German, and Italian. How can I help?")

    return f"[{detected_lang.upper()}]: {response}"
```

### Chatbot Demonstration

Let's test the chatbot with inputs in different languages.
